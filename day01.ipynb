# Part 1

```python
import csv
values_list = []
with open("data/day01.csv", 'r') as file:
  csvreader = csv.reader(file)
  for row in csvreader:
    values_list.append(row)

values_list
```

```python
meal_sums = []
sum=0

for item in range(len(values_list)):
    if len(values_list[item]) == 1:
        sum += int(values_list[item][0])
    else:
        meal_sums.append(sum)
        sum=0

max(meal_sums)
```

```python
sum_of_3 = 0
for i in range(3):
    print(max(meal_sums))
    sum_of_3 += max(meal_sums)
    meal_sums.remove(max(meal_sums))

sum_of_3
```
