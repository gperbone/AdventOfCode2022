```python
with open("data/day10.txt") as f:
    lines = f.read().splitlines()
```

```python
X=1
cycle = 0

list_record = [] # [cycle, value]

for line in lines:
    command = line.split(" ")
    if command[0] == "noop":
        cycle += 1
        list_record.append([cycle, X])
    else:
        list_record.append([cycle+1, X])
        list_record.append([cycle+2, X])
        cycle += 2
        X += int(command[1])

sum = 0
for item in list_record:
    if item[0] in [20, 60, 100, 140, 180, 220]:
        print(item[0], item[1])
        sum += item[0] * item[1]

sum
```
