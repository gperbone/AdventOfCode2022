```python
with open("data/day07.txt") as f:
    lines = f.read().splitlines()
```

```python
class File:
    def __init__(self, name, size):
        self.name = name
        self.size = size


class Directory:
    def __init__(self, name, parent):
        self.name = name
        self.parent = parent
        self.content = {}
        self.size = 0
    
    def add_file(self, file_name, size):
        self.content[file_name] = File(file_name, size)
        self.size += int(size)

    def add_directory(self, dir):
        self.content[dir] = Directory(dir, self)


root = None
parent_dir = None
curr_dir = None
add_mode = False

for line in lines:
    if line[2:4] == 'cd':
        add_mode = False

        if curr_dir == None:
            root = Directory(line[5:], None) 
            curr_dir = root
        elif line[5:] == '..':
            curr_dir = curr_dir.parent
        else:
            curr_dir = curr_dir.content[line[5:]]

    if add_mode:
        if line.split()[0] == 'dir':
            dir = line.split()[1]
            curr_dir.add_directory(dir)
        
        else:
            curr_dir.add_file(line.split()[1], line.split()[0])  

    if line[2:4] == 'ls':
        add_mode = True   

print(root)
```

```python
def sum_dir_size(directory_root, path, size_total):
    size_total[path] = 0

    for key,value in directory_root.content.items():
        if isinstance(value, Directory):
            size_total[path] += sum_dir_size(value, f'{path}{key}/', size_total)

    size_total[path] += directory_root.size

    return size_total[path]

size_total = {}
sum_dir_size(root, root.name, size_total)

size_total
```

```python
sum = 0

for _, v in size_total.items():
    if v <= 100000:
        sum += v

sum
```
