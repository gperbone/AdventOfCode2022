```python
with open("data/day14.txt") as f:
    lines = f.read().splitlines()

infos = [line.replace(" ","").split("->") for line in lines if line != ""]
infos
```

```python
import numpy as np

cave = {}

for info in infos:
    coords = [(int(item.split(',')[0]), int(item.split(',')[1])) for item in info]

    for p1, p2 in zip(coords, coords[1:]):
        x1, y1 = p1
        x2, y2 = p2

        for x in range(min(x1, x2), max(x1, x2) + 1):
            cave[(x, y1)] = "#"
 
        for y in range(min(y1, y2), max(y1, y2) + 1):
            cave[(x1, y)] = "#"
```

```python
def movement_of_sand(cave):
    x = 500

    for y in range(max(y for (_,y) in cave.keys())):
        if (x, y+1) not in cave:
            continue
        # fall to the left
        elif (x - 1, y + 1) not in cave:
            x -= 1
        # fall to the right
        elif (x + 1, y + 1) not in cave:
            x += 1
        else: 
            cave[(x, y)] = "o"
            # stop simulation if we reach the initial point
            return (x, y) != (500, 0) 
    return False 
 
def start_falling_sand(cave):
    while movement_of_sand(cave):
        pass
    return cave

sum(value == "o" for value in start_falling_sand(cave).values())

```

```python
cave_floor_y = max(y for (_,y) in cave.keys()) + 2
smallest_x = min(x for (x,_) in cave.keys()) - 1000 # 100 to represent infinity floor
largest_x = max(x for (x,_) in cave.keys()) + 1000

for x in range(smallest_x, largest_x):
    cave[(x,cave_floor_y)] = "#"

sum(value == "o" for value in start_falling_sand(cave).values())
```
