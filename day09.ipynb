```python
with open("data/day09.txt") as f:
    lines = f.read().splitlines()
```

```python
import numpy as np
moves = lines
moves
```

```python
def move_head(H, direction):
    if direction == 'U':
        H[0] -= 1
    if direction == 'D':
        H[0] += 1
    if direction == 'L':
        H[1] -= 1
    if direction == 'R':
        H[1] += 1

    return H

def move_tail(H,T):
    if abs(H[0]-T[0])+abs(H[1]-T[1]) == 3:
        if H[0]==T[0]-2:
            T[0]-=1
            T[1]+=H[1]-T[1]
        if H[0]==T[0]+2:
            T[0]+=1
            T[1]+=H[1]-T[1]
        if H[1]==T[1]-2:
            T[1]-=1
            T[0]+=H[0]-T[0]
        if H[1]==T[1]+2:
            T[1]+=1
            T[0]+=H[0]-T[0]

    elif abs(H[0] - T[0]) > 1 or abs(H[1] - T[1]) > 1:
        if H[0]==T[0]-2:
            T[0]-=1
        if H[0]==T[0]+2:
            T[0]+=1
        if H[1]==T[1]-2:
            T[1]-=1
        if H[1]==T[1]+2:
            T[1]+=1
    return T
```

```python
H = np.array([0,0])
T = np.array([0,0])
passed = set([tuple(T)])

for move in moves:
    direction = move.split(" ")[0]
    number = int(move.split(" ")[1])
    while number > 0:
        H = move_head(H, direction)
        T = move_tail(H, T)
        passed.add(tuple(T))
        number -= 1



len(passed)
```
